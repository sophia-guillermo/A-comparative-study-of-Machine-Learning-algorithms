### 3.1 Decision Tree (DT)

#### 3.1.1 How DTs make decisions
This algorithm is similar to a guessing game in which we are asked questions until we get the answer right. The root node, which has the broadest question is the node located at the very top of tree. As we progress down the tree's levels, the questions get more detailed. The final answer, is known as the leaf node, gets approached more closely, the more questions we ask.

Likewise, the DT algorithm bases its questions on the features of the data. For example, a question in the chosen Wine dataset could go like "alcohol <= 13.0" or "proline >= 1065". The tree would divide the feature space into subsets at each node, to determine which of the 13 features to ask about.

#### 3.1.2 Advantages of DTs
Applying DTs would have the advantages of interpretability and robustness to outliers. For interpretability, since DTs are presented in a straightforward manner, in a form of a flowchart, it is easy to follow and comprehend why a certain prediction was made. As for its robustness to outliers, given that its choices are based on majority of the data, having unusual data points, do not mess up the decisions made by the DTs.

#### 3.1.3 Potential Disadvantages of DTs
Applying DTs do come with its disadvantages such as overfitting and biasness. If there is excessive splitting this would result in overfitting of data as the model would fit the training data too closely, making it bad at predicting data it has not seen before. This is similar to memorizing the answers to an exam, instead of understanding its underlying concepts.

### 3.2 Logistic Regression (LR)
I would explain the analogy of LR using the example of trying to predict whether a person would enjoy a movie, based on specific features such as genres, ratings and casts.

#### 3.2.1 Modeling the relationship
This is similar to trying to determine how these features relate to the possibility of a person enjoying the movie. The LR model utilizes a sigmoid function on top of a linear function to keep the predictions between 0 and 1.

#### 3.2.2 Estimating probabilities
With that, suppose a person is thinking about watching a new horror movie. Using the linear function, the LR model would compute a value which would be passed into the sigmoid curve. There is an 70% likelihood that the person would like the horror movie if the outcome is 0.7. If the outcome is more than 0.5, the LR model would predict that the person is liekly going to find the movie enjoyable. Otherwise, it would suggest they might not be thrilled to watch the movie

#### 3.2.3 Advantages of Logistic Regression
Just like DTs, LRs are also easy to interpret. It is the same as discussing preferences with a friend. Similar to saying, <i>"Your probability of liking a movie increases by 5% for every extra jumpscare scene."</i>. We are able to comprehend and illustrate the significance of every feature because of its interpretability. Futhermore, LR works effectively on bigger datasets. 

#### 3.2.4 Potential disadvantages Logistic Regression
However, LR does have its drawbacks such as assuming the relationship between the said movie features and enjoyment as a straight line, which tends to oversimplify things. This is like making the assumption that the more jumpscares there are, the higher chance of enjoyment, which tends to overcomplicate things. Another disadvantage of LR would be its unsuitability for complex and non-linear relationships.

### 3.3 K-Nearest Neighbors (KNN)
I would explain the analogy of KNN using the example of trying to figure out which group of friends to hangout with in a party. 

#### 3.3.1 Concept of Distance Metrics
Distance metrics is like having a measuring tape, to know how close (in terms of distance) we are from a group of friends. The closer we are to them, the more we can vibe with one another in the party. In KNN, Euclidean distance shows how close data points are in terms of their features. The closer their distance is, the more similar their data points would be. 

#### 3.3.2 KNN Classification 
Next, KNN is similar to asking people around their interest to know who we are most similar to. If we want to know if a person has a hobby of swimming or reading, we look at the hobbies of their nearest friends. If most of them have the hobby of reading, we would most likely have the hobby of reading too. KNN does the same by checking the majority class (hobby of reading or swimming) among our <i>k</i> nearest friends.

#### 3.3.3 Choosing the Right <i>k</i>
<i>k</i> in KNN is like choosing how many people we would want to ask their interests for. If we ask too few people, like 1 to 2, we might get influenced by their preferred hobby. Whereas if we ask too many people like 100 to 150 people, it will become to chaotic and difficult in selecting our choice of hobby. 
It is important to choose the right <i>k</i>, if it is too small it would lead to overreacting to noise and too large would make us miss out on important details since KNN would get confused by the diverse data, making inaccurate predictions. 

#### 3.3.4 Potential Disadvantages of KNN
For big datasets, calculating the distances for every point would be extremely slow, which is comparable to waiting in line to get into the venue of the party. Another disadvantage would be having irrelevant features. If we have friends who are into reading but we ask them about their height, we would be tricked into choosing the wrong group because it considered this very feature.

#### 4.3.4 Implementing KNN from scratch
In this section, I would be elaborating on how I implemented KNN from scratch and likewise train the model using the training data.

My implementation involves defining a function that iterates through each test data point. For each of the test point, the function would calculate the distances to training points, based on the specified metric. Next, the k-nearest neighbors would be identified, then, voting would be performed to predict the label based on the majority vote. These predictions would be stored in an array. The entire process would be repeated for all test data, returning it in a list containing all the predicted labels.

Each line of code would be explained, as shown below.

[1] KNN_algorithm function is defined, taking in four parameters:
1. Features of training data
2. Labels of training data
3. Features of test data
4. Number of neighbours to consider
5. Type of distance metric to use

[2] An empty list called 'preds' is initialized to store all the predicted label for each point in the test data.

[3] A for loop to iterate through each point in the test data 'X_test'

[4] Another empty list called 'dists' is initialized, this time storing the distances between a point in the test data and all the points in the training data.

[5] A nested for loop to iterate through each point in the training data 'X_train'

[6-11] Handles the distance metrics computation based on the parameter passed in. If 'manhattan' was chosen, the necessary distance calculation between test_data_point and train_data_point would be calculated. Likewise, if 'euclidean' was chosen, the necessary distance calculation would be computed. Otherwise, an error would be thrown if the given distance metric is neither of the two.

[12] The calculated distance, based on the metric chosen is appended to the list 'dists' in line 4.

[13] Line 13 finds the indices of the k-nearest neighbors, by sorting the distances and selecting the first k indices. The NumPy function `np.argsort` returns the indices that would sort the array. 

[14] Another empty list called 'labels_of_neighbors' is initialized. It stores the labels of the k-nearest neighbors.

[15] One more nested for loop is created to iterate through the indices of the k-nearest neighbors and retrieve their corresponding labels from the array of labels in the training data, 'y_train'

[16] The corresponding labels are then appended to the list 'labels_of_neighbors' in line 14.

[17] The labels are converted into a set so that only unique labels are returned. `max()` finds the maximum element in the set, based on the `key`. Which in this case is specified as 'labels_of_neighbors.count', which returns the count of occurrences of each label. As the 'max' function iterates through the unique labels in the set, the label with the maximum count determines the predicted label for the current point in the test data.

[18] The majority vote derived from line 18 is appended to the list of 'preds' in line 2.

[19] Finally, the function would return the list of predicted label for all the points in the test data.

```python
def KNN_algorithm(X_train, y_train, X_test, k, metric='manhattan'):
    preds = []
    for test_data_point in X_test:
        dists = []
        for train_data_point in X_train:
            if metric == 'manhattan':
                distance = np.sum(np.abs(test_data_point - train_data_point))
            elif metric=='euclidean':
                distance = np.sqrt(np.sum((test_data_point - train_data_point)**2))
            else:
                raise ValueError("The distance metric specified does not exists. Choose between'euclidean' or 'manhattan'.")
            dists.append(distance)

        indices_of_neighbors = np.argsort(dists)[:k]
        labels_of_neighbors = []
        for i in indices_of_neighbors:
            labels_of_neighbors.append(y_train[i])
        
        majority_vote = max(set(labels_of_neighbors), key=labels_of_neighbors.count)
        preds.append(majority_vote)
    
    return preds
```

```python
knn_preds = KNN_algorithm(X_train, y_train, X_test, k=7)
```

#### 5.2.3 CV for KNN from scratch

The `cross_val_score` expects an object that implements the `fit` method. However, since the KNN algorithm was implemented from scratch, I do not have a `fit` method implemented. Therefore, a k-fold cross validation for my KNN algorithm must be done from scratch.

Bhattacharjee (2023) provides a clear example of implementing k-fold cross-validation, as shown in the following code. I have changed parts of the code to fit my `KNN_algorithm` into the cross validation.

```python
def knn_cross_val(X, y, k, metric='manhattan'):
    kf = KFold(n_splits=5, shuffle=True)
    scores = []

    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        knn_preds = KNN_algorithm(X_train, y_train, X_test, k,metric='manhattan')
        fold_score =  accuracy_score(y_test, knn_preds)
        scores.append(fold_score)

    mean_accuracy = np.mean(scores)
    return scores, mean_accuracy

k = 7
knn_cv_accuracies, knn_cv_mean_accuracy = knn_cross_val(X, y, k,metric='manhattan')
print("KNN Cross-Validation Scores:", knn_cv_accuracies)
print("Mean CV Accuracy of KNN from sratch):", knn_cv_mean_accuracy)
```

### 5.3 Hyperparameter Tuning
This section includes subsections for hyperparameter tuning for each model. The tuning method I chose to implement is Grid Search and my evaluation metrics remain the same as per the previous section, of which are accuracy score, precision score, recall score and F1-score. In each sub-section, would be the goal I am trying to achieve and the chosen hyperparameters.

#### 5.3.1 Decision Tree Hyperparameter Tuning
The goal for the DT model is to have a balanced model complexity so that it would not overfit data. In the codes below, I aim to find the optimal tree structure to use, so I could pass it in the retrained DT model in section 6.1 Retraining all models after hyperparameter tuning.

As for the LR model, my aim is to control the regularization so as to reduce overfitting by adjusting the complexity of the model.

The hyperparameters tuned below are `C` which indicates the strength of regularization and `penalty`.

```python
lr_parameters_grid = {
    'C': [0.01, 0.1, 0.5, 1.0],
    'penalty': ['l2']
}

lr_grid_search = GridSearchCV(estimator=lr, param_grid=lr_parameters_grid, cv=5)

lr_grid_search.fit(X_train, y_train)

lr_best_parameters = lr_grid_search.best_params_

print("Best hyperparameters for Logistic Regression: ", lr_best_parameters)
```

#### 5.3.3 KNN Hyperparameter Tuning
The goal for the KNN model is to determine the optimal number of neighbors to use and explore other distance metrics, to see if it would perform better.

The hyperparameters tuned below are `k` and `metric`.

```python
knn_best_score = 0
knn_best_parameters = None

parameters_grid_KNN = {
    'k': [3,5,7],
    'metric': ['euclidean', 'manhattan']
}

for params in ParameterGrid(parameters_grid_KNN):
    k = params['k']
    metric = params['metric']

    scores, mean_accuracy = knn_cross_val(X, y, k, metric)

    if mean_accuracy > knn_best_score:
        knn_best_score = mean_accuracy
        knn_best_parameters = params

print("Best hyperparameters for K-Nearest Neighbors: ", knn_best_parameters)
```
